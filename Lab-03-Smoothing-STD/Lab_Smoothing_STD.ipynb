```python
'''
(1) Your own median filter algorithm
'''
def custom_median_filter(image, window_size=3):
  output_image = np.zeros(image.shape)
  
  edge_x = window_size // 2
  edge_y = window_size // 2
  
  for i in range(edge_x, image.shape[0] - edge_x):
    for j in range(edge_y, image.shape[1] - edge_y):
      window = image[i-edge_x:i+edge_x+1, j-edge_y:j+edge_y+1]
      output_image[i, j] = np.median(window)
  
  return output_image
```

```python
'''
(2) skimage median filter
'''
image = io.imread('pout.tif')

image_with_noise = random_noise(image, mode='s&p', amount=0.05)

custom_median_filtered_image = custom_median_filter(image_with_noise, window_size=3)
skimage_median_filtered_image = median(image_with_noise)

show_images([image, image_with_noise, custom_median_filtered_image, skimage_median_filtered_image], ['Original Image', 'Image with Noise', 'Custom Median Filtered Image', 'Skimage Median Filtered Image'])
```

```python
'''
(3) Gaussion Filters
'''
gaussian_sigma_8 =  gaussian(image_with_noise, sigma=8)
gaussian_sigma_3 =  gaussian(image_with_noise, sigma=3)
gaussian_sigma_02 =  gaussian(image_with_noise, sigma=0.2)

show_images([gaussian_sigma_8, gaussian_sigma_3, gaussian_sigma_02], ['Gaussian Filtered Image (Sigma=8)', 'Gaussian Filtered Image (Sigma=3)', 'Gaussian Filtered Image (Sigma=0.2)'])

gaussian_sigma_2 =  gaussian(image_with_noise, sigma=2)
gaussian_sigma_1 =  gaussian(image_with_noise, sigma=1)
gaussian_sigma_05 =  gaussian(image_with_noise, sigma=0.5)

show_images([gaussian_sigma_2, gaussian_sigma_1, gaussian_sigma_05], ['Gaussian Filtered Image (Sigma=2)', 'Gaussian Filtered Image (Sigma=1)', 'Gaussian Filtered Image (Sigma=0.5)'])

gaussian_sigma_23 =  gaussian(image_with_noise, sigma=2.3)
gaussian_sigma_25 =  gaussian(image_with_noise, sigma=2.5)
gaussian_sigma_27 =  gaussian(image_with_noise, sigma=2.7)

show_images([gaussian_sigma_23, gaussian_sigma_25, gaussian_sigma_27], ['Gaussian Filtered Image (Sigma=2.3)', 'Gaussian Filtered Image (Sigma=2.5)', 'Gaussian Filtered Image (Sigma=2.7)'])
```
